Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Answer(2):

In Bokeh, glyphs are geometric shapes or markers that we can use to represent data points on a plot. Glyphs are a fundamental building block of Bokeh plots, and It can be added to a plot to visualize our data. Common glyphs include circles, squares, lines, and more.

Here's how we can add glyphs to a Bokeh plot with an example using the circle glyph:

```python
'''Answer(2)'''
from bokeh.plotting import figure, show

# Create a new Bokeh plot
p = figure(title='Line_plot',width=600, height=400)

# Add a circle glyph to the plot
x = [9,8,7,6,5,4]
y = [1,2,3,4,5,6]
p.circle(x, y,color="red",size=10,alpha=0.4)    #here circle is used red and size and for transparency for circle alpha parameter is used 0.4

# Show the plot
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Answer(4):

A Bokeh server is a component of the Bokeh library that allows us to create interactive web applications and dashboards with real-time updates. With the Bokeh server, We can build dynamic data visualizations that respond to user interactions, data changes, or external inputs, providing a rich and interactive user experience.

Here's an overview of how to use a Bokeh server to create interactive plots that can be updated in real time:

```python
'''Answer(4):'''
from bokeh.models import ColumnDataSource
from bokeh.plotting import curdoc, figure
from bokeh.models.widgets import Slider

source = ColumnDataSource(data={'x': [1, 2, 3, 4, 5, 6], 'y': [1, 2, 3, 4, 5, 6]})

plot = figure(width=400, height=400)
plot.circle('x', 'y', source=source)

slider = Slider(start=0, end=10, value=5, step=0.1, title="Slider")

def update_data(attrname, old_val, new_val):
    # Update the data source based on the slider value
    x = range(int(slider.value))
    y = [i**2 for i in x]
    source.data = {'x': x, 'y': y}

# Attach the callback function to the slider's value attribute
slider.on_change('value', update_data)

curdoc().add_root(slider)
curdoc().add_root(plot)


bokeh serve my_bokeh_app.py #to run tihs code

```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Answer(5):
Embedding a Bokeh plot into a web page or dashboard using Flask or Django is a great way to integrate interactive data visualizations into our web applications. Both Flask and Django are popular Python web frameworks that can be used for this purpose. Here are the general steps for embedding a Bokeh plot into a web page or dashboard in each framework



```python
'''Answer(5)'''

from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components
from bokeh.models import ColumnDataSource
from bokeh.models.widgets import Slider

app = Flask(__name__)

@app.route('/')
def index():
    # Create a simple Bokeh plot
    source = ColumnDataSource(data={'x': [1, 2, 3, 4, 5, 6], 'y': [1, 2, 3, 4, 5, 6]})
    plot = figure(width=400, height=400)
    plot.circle('x', 'y', source=source)

    # Create a slider widget
    slider = Slider(start=0, end=10, value=5, step=0.1, title="Slider")

    def update_data(attrname, old_val, new_val):
        # Update the data source based on the slider value
        x = range(int(slider.value))
        y = [i**2 for i in x]
        source.data = {'x': x, 'y': y}

    # Attach the callback function to the slider's value attribute
    slider.on_change('value', update_data)

    # Generate Bokeh components for embedding
    script, div = components(plot)

    # Render the HTML template and pass the script and div components
    return render_template('index.html', script=script, div=div, slider=slider)

if __name__ == '__main__':
    app.run(debug=True)
```
