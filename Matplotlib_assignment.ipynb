Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer-1
Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in 2D and 3D. It provides a wide range of tools for constructing various types of plots and charts, making it a valuable tool for data visualization and scientific plotting.

Matplotlib is used for several purposes, including:

Data Visualization: Matplotlib is primarily used for creating a wide variety of data visualizations, helping users to explore and communicate their data effectively.

Publication-quality Plots: It allows users to create high-quality plots suitable for publications and presentations with fine-grained control over every aspect of the plot.

Customization: Matplotlib offers extensive customization options, allowing users to tailor the appearance of their plots to meet specific requirements.

Interactivity: It supports interactive features for zooming, panning, and annotating plots, which can be useful for data exploration.

Integration: Matplotlib can be integrated into various Python environments and libraries, making it a versatile tool for creating plots in data analysis workflows.

Here are five common types of plots that can be created using the pyplot module of Matplotlib:

Line Plot: Line plots are used to represent data points connected by lines, typically used for time series data or showing trends over continuous variables.

Scatter Plot: Scatter plots display individual data points as markers on a 2D plane, often used to visualize the relationship between two continuous variables.

Bar Plot: Bar plots (bar charts) are used to represent categorical data with rectangular bars. They are useful for comparing values across different categories.

Histogram: Histograms show the distribution of a single variable by dividing it into bins and counting the number of data points in each bin. They are useful for understanding the data's underlying distribution.

Pie Chart: Pie charts represent data as sectors of a circle, with each sector's size proportional to the data it represents. They are often used to display the composition of a whole, showing parts as percentages.

Matplotlib provides a wide range of other plot types as well, including heatmaps, box plots, violin plots, and more, making it a versatile library for data visualization in python

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np

Note: Also add title, xlabel, and ylabel to the plot.

Answer-2
A scatter plot is a type of data visualization that is used to display individual data points in a two-dimensional space. In a scatter plot, each data point is represented as a marker or point on a Cartesian coordinate system, with one variable plotted along the x-axis and another variable plotted along the y-axis. These plots are particularly useful for visualizing the relationship or correlation between two continuous variables.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y,c="green")
plt.xlabel("<------------X-axis------------->")
plt.ylabel("<------------Y-axis------------->")
plt.title('Scatter diagram')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


Ans-The subplot function in Matplotlib is used to create a grid of subplots within a single figure. Each subplot is essentially a smaller plot within the larger figure, allowing you to display multiple plots or visualizations side by side or in a grid arrangement. This function is particularly useful when you want to compare and visualize different aspects of your data simultaneously.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2x2 grid of subplots
plt.figure(figsize=(10, 8))  # Sets the overall figure size
plt.subplots_adjust(hspace=0.4)  # Adjusts vertical spacing between subplots

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Activating the first subplot
plt.plot(x1, y1, marker='o', linestyle='-', color='b')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Activating the second subplot
plt.plot(x2, y2, marker='s', linestyle='--', color='g')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Activating the third subplot
plt.plot(x3, y3, marker='^', linestyle=':', color='r')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Activating  the fourth subplot
plt.plot(x4, y4, marker='*', linestyle='-.', color='purple')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Displays the figure with all subplots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans-A bar plot, also known as a bar chart or bar graph, is a type of data visualization that uses rectangular bars or columns to represent data. Bar plots are particularly useful for displaying and comparing the values of different categories or groups of data. They are commonly used in data analysis, statistics, and business reporting to illustrate patterns, trends, and comparisons.
 
Uses of bar plots include:

Categorical Data Representation: Bar plots are best suited for displaying categorical data, where data points are organized into distinct categories or groups. These categories can be anything from product names, months, countries, to any discrete classification.

Bar Length Represents Data Values: In a bar plot, the length or height of each bar is proportional to the value it represents. The longer the bar, the larger the value it represents.

Comparisons: Bar plots are excellent for comparing values across different categories or groups. They make it easy to identify which category has the highest or lowest value, as well as to see relative differences between categories.

Variations: There are various types of bar plots, including vertical bar plots (where bars are oriented vertically), horizontal bar plots (where bars are oriented horizontally), grouped bar plots (for comparing values within subcategories), and stacked bar plots (for showing the composition of a whole as well as individual parts).

Data Labels: Bar plots often include data labels or values at the top or end of each bar to provide precise information about each category's value.

Customization: Bar plots can be customized to adjust the color of bars, add legends, change bar widths, and more to make the visualization more informative and visually appealing.

```python
'''Answer(4)'''

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit,width=0.4)
plt.title("Bar plot")
plt.show()
```

```python
'''Answer(4)'''
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit,height=0.4)
plt.title(" Horizontal bar plot")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans-A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution of a dataset and provides a summary of its key statistical characteristics, such as the median, quartiles, and potential outliers. Box plots are useful for understanding the spread and central tendency of a dataset and for identifying any skewness or extreme values.

```python
'''Answer(5)'''
import numpy as np
import matplotlib.pyplot as plt

# Generate the data for the box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot with two boxes
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add labels and a title
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')


plt.show()
```
