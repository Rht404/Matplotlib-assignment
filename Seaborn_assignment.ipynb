Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans-Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots that you can create using Seaborn, along with their typical use cases:

1.Scatter Plot
    Use: Scatter plots are used to visualize the relationship between two numerical variables. They are helpful in identifying patterns, clusters, or correlations in data.

2.Histogram
    Use: Histograms display the distribution of a single numerical variable by dividing it into bins and counting the frequency of data points in each bin. They help in understanding the data's underlying distribution.

3.Bar Plot
    Use: Bar plots are useful for visualizing categorical data, showing comparisons between different categories. They can be used to display counts, sums, or averages of data within each category.

4.Box Plot (Box-and-Whisker Plot)
    Use: Box plots provide a summary of the distribution of numerical data. They show the median, quartiles, and identify potential outliers, making them effective for identifying skewness or variability in data.

5.Heatmap
    Use: Heatmaps are used to visualize the relationships between two categorical variables by displaying them in a grid format. They are particularly useful for exploring correlation matrices or showing the concentration of data in different categories.

Seaborn also provides support for more advanced plots like pair plots, violin plots, and regression plots, among others, to assist in various data analysis and visualization tasks.

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')   #loading the flights dataset 
```

```python
flights
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset from Seaborn
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")  # annot=True displays values, fmt="d" for integer formatting, cmap for color map
plt.title("Passenger Count Heatmap (Flights Dataset)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
